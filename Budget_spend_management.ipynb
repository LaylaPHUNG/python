```python
# Create a new DataFrame with the necessary columns
df_expense = df[['expense_category', 'Budget', 'spends']]

# Create a pivot table with expense_category as the index and Budget and Spends as columns
pivot_df = pd.pivot_table(df_expense, index='expense_category', values=['Budget', 'spends'])

# Display the pivot table
print(pivot_df)
```

```python
df2 = pivot_df.reset_index()
```

```python
df2
```

```python
category = df2['expense_category']
budget = df2['Budget']
spends = df2['spends']

fig, ax = plt.subplots()

b1 = ax.barh(category, budget, color="blue")
b2 = ax.barh(category, spends, left=budget, color="orange")

ax.set_xlabel('Amount')
ax.set_ylabel('Category')
ax.set_title('Budget vs Spends by Expense Category')
ax.legend([b1, b2], ["Budget", "Spends"], loc="upper right")


```
